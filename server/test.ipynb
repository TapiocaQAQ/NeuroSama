```python
def convert_messages_to_prompt(messages):
    """
    Convert a list of message dictionaries to a formatted prompt string.
    
    Args:
    messages (list): A list of dictionaries with 'role' and 'content' keys
    
    Returns:
    str: Formatted prompt string
    """
    formatted_prompt = system_prompt = '''<|begin_of_text|><|start_header_id|>system<|end_header_id|>\n\nCutting Knowledge Date: December 2023\n\n'''
    
    
    # Process user and assistant messages
    for message in messages:
        if message['role'] in ['user', 'assistant']:
            formatted_prompt += (
                f"<|start_header_id|>{message['role']}<|end_header_id|>\n"
                f"{message['content']}"
                f"<|eot_id|>\n"
            )
    
    return formatted_prompt
```
